````python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Datos
x = [0, 100]
y = [0, 200]

# Crear el gr√°fico principal
fig, ax = plt.subplots()
ax.plot(x, y, 'r')  # L√≠nea roja
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Ejercicio 2')

# Crear el gr√°fico insertado
ax_inset = inset_axes(ax, width="20%", height="20%", loc='upper center')
ax_inset.plot(x, y, 'r')  # L√≠nea roja
ax_inset.set_xlim(0, 100)
ax_inset.set_ylim(0, 200)
ax_inset.set_xticks([0, 20, 40, 60, 80, 100])
ax_inset.set_yticks([0, 50, 100, 150, 200])

# Mostrar el gr√°fico
plt.show()
````

````python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import numpy as np

# Datos
x = np.linspace(0, 100)
y = np.exp(x / 100)

# Crear el gr√°fico principal
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_title('Ejercicio 3')

# Crear el gr√°fico insertado
ax_inset = inset_axes(ax, width="30%", height="30%", loc='upper left')
ax_inset.plot(x, y)
ax_inset.set_xlim(20, 22)
ax_inset.set_ylim(30, 45)
ax_inset.set_xticks([20, 21, 22])
ax_inset.set_yticks([30, 35, 40, 45])

# Mostrar el gr√°fico
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Datos para el gr√°fico azul
x1 = np.linspace(0, 100, 100)
y1 = 2 * x1

# Datos para el gr√°fico rojo
x2 = np.linspace(0, 100, 100)
y2 = np.exp(x2 / 10)

# Crear el gr√°fico azul
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 6))

ax1.plot(x1, y1, 'b--')  # L√≠nea azul discontinua
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Gr√°fico Azul')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 200)

# Crear el gr√°fico rojo
ax2.plot(x2, y2, 'r')  # L√≠nea roja continua
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Gr√°fico Rojo')
ax2.set_xlim(0, 100)
ax2.set_ylim(0, 10000)

# Mostrar los gr√°ficos
plt.suptitle('Ejercicio 4')
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Ejes
fig = plt.figure()
ax = plt.axes(projection='3d')

# L√≠nea
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'blue')

# Nube de puntos
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Reds')

# Mostrar el gr√°fico
plt.show()
````
